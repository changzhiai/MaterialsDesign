```python
def energy(x, y):
    z = np.log((2*(y-x **2))**2 + (1 - x)**2 + 1)
    return z
def force(x, y):
    dx = (-16*x*(y-x**2)+2*x - 2)/(4*(y-x**2)**2 + (1 - x)**2 + 1)
    dy = (8*(y - x**2))/(4*(y-x**2)**2 + (1 - x)**2 + 1)
    return dx, dy
```
